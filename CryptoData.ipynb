# Cryptocurrency Clusters

```python
# Initial imports
import pandas as pd
from pathlib import Path
```

## Data Preparation

```python
# Data loading
file_path = Path("Resources/crypto_data.csv")
df_crypto = pd.read_csv(file_path)
df_crypto
```

```python
# Discard all cryptocurrencies that are not being traded
df_crypto = df_crypto[df_crypto.IsTrading !=False]
df_crypto
```

```python
# Drop the IsTrading column from the dataframe
df_crypto = df_crypto.drop(columns=["IsTrading"])
df_crypto
```

```python
# Remove all rows that have at least one null value
df_crypto.dropna(inplace=True)
df_crypto
```

```python
#Filter for cryptocurrencies that have been mined. The total coins mined should be greater than zero.
df_crypto['TotalCoinsMined'] = df_crypto[df_crypto['TotalCoinsMined'] > 0]['TotalCoinsMined']
df_crypto.dropna(inplace=True)
df_crypto
```

```python
#Delete the CoinName column bc not numeric or important
df_crypto = df_crypto.drop(columns=["CoinName"])
df_crypto
```

```python
crypto_df_index=df_crypto[["Unnamed: 0","Algorithm","ProofType","TotalCoinsMined","TotalCoinSupply"]].copy().set_index("Unnamed: 0")
crypto_df_index
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df = crypto_df_index.copy()
df['Algorithm'] = le.fit_transform(df['Algorithm']) 
df['ProofType'] = le.fit_transform(df['ProofType']) 
df
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

## Dimensionality Reduction

```python
# Initialize PCA model
from sklearn.decomposition import PCA

pca = PCA(n_components=0.90)

# Get principal components for the data
crypto_pca = pca.fit_transform(scaled_data)
```

```python
# Fetch the explained variance
pca.explained_variance_ratio_
```

```python
from sklearn.manifold import TSNE
# Initialize t-SNE model
tsne = TSNE(learning_rate=100)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(crypto_pca)
```

```python
tsne_features.shape
```

```python
# Prepare to plot the dataset
# The first column of transformed features
df['x'] = tsne_features[:,0]

# The second column of transformed features
df['y'] = tsne_features[:,1]
```

```python
# Visualize the clusters
import matplotlib.pyplot as plt

plt.scatter(df['x'], df['y'])
plt.show()
```

## Cluster Analysis with k-Means

```python
from sklearn.cluster import KMeans

# Finding the best value for k
inertia = []
k = list(range(1, 10))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(10)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

## Recommendation

Based on my findings, the cryptocurrencies can be clustered together into about 4 clusters. The t-SNE scatterplot showed a couple of distinct cryptocurrency clusters, while the k-Means graph showed that after 4 clusters, the amount of inertia doesn't vary as much.
